```python
import numpy as np  #import the fundamental package for scientific computing with Python
import matplotlib.pyplot as plt #import the basic library for plotting 2D-graphs

from sklearn import linear_model #Import a set of methods intended for regression in which 
#the target value is expected to be a linear combination of the input variables.
from sklearn import datasets #scikit-learn comes with a few standard datasets
```

```python
diabetes = datasets.load_diabetes() #load the diabetes dataset
X = diabetes.data #separation of feature variables
y = diabetes.target #instantiate the target variable
print(diabetes.data) 
```

```python
print("Computing regularization path using the LARS ...")
alphas, _, coefs = linear_model.lars_path(X, y, method='lars', verbose=True) #Compute Least 
#Angle Regression or Lasso path using LARS algorithm

xx = np.sum(np.abs(coefs.T), axis=1) #Sum of the absolute value of array elements over a given axis;
xx /= xx[-1] #ratio of entries in the features 

plt.plot(xx, coefs.T) #plot the lars path
ymin, ymax = plt.ylim() #Setting limits for the y-axis on autoscaling  
plt.vlines(xx, ymin, ymax, linestyle='dashed') #Plot vertical lines at each x from ymin to ymax.
plt.xlabel('|coef| / max|coef|') #labeling the x-axis
plt.ylabel('Coefficients') #labeling the y-axis
plt.title('LARS Path')  #labeling the diagram itself
plt.axis('tight') #The tight setting is retained for future autoscaling until it is explicitly changed.
plt.show() #draw the plot
```
